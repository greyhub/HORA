```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.neighbors import NearestNeighbors
```

# Data

```python
df = pd.read_csv('../data/cleaned-table/basic_info.csv')
df
```

```python
df.set_index('hotel_id')
```

# Training

```python
nbrs = NearestNeighbors(n_neighbors=5, algorithm='ball_tree').fit(df.set_index('hotel_id'))
```

> Testing

```python
test = df.set_index('hotel_id').values[0].reshape(1,-1)
nbrs.kneighbors(test, return_distance=False)
```

# Saving Model

```python
import pickle 

knn = NearestNeighbors(n_neighbors=5, algorithm='ball_tree').fit(df.set_index('hotel_id'))

# Its important to use binary mode 
knnPickle = open('knnpickle_file', 'wb') 

# source, destination 
pickle.dump(knn, knnPickle)                      
```

# Loading Model

```python
# load the model from disk
loaded_model = pickle.load(open('knnpickle_file', 'rb'))
```

```python
df.set_index('hotel_id').values[0].reshape(1,-1)
```

```python
# input
hotel_id = 39339
# load data-to-model
hotel_info = df[df['hotel_id'] == 39339].index[0]
# check
test = df.set_index('hotel_id').values[hotel_info].reshape(1,-1)
result = loaded_model.kneighbors(test, return_distance=False)
result
```
