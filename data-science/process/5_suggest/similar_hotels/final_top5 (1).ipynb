```python
import pandas as pd
import numpy as np
```

```python
joined = pd.read_csv("../data/joined.csv")
joined.T
```

```python
pivot = pd.pivot_table(joined, index='hotel_id', columns='user_id', values='overall_score')
pivot.head()
```

```python
def normalizeData(x):
    x =(pivot-pivot.mean())/pivot.std()
```

```python
from sklearn import preprocessing

x = pivot.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
```

```python
df.fillna(0, inplace=True)
df.head().value_counts()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
item_sim_df = pd.DataFrame(cosine_similarity(df, df), index=df.index, columns=df.index)
item_sim_df.head()
```

```python
# item_sim_df.to_csv("../data/dfSimilarPoint.csv")
```

```python
hotel_id_5127 = item_sim_df.nlargest(6, [5127]) 
hotel_id_5127 = hotel_id_5127[1:]
hotel_id_5127
```

```python
hotel_id_top5 = {}
for i in range(100):
    top5 = []
    top5 = item_sim_df.nlargest(6,[i])
    top5 = top5[1:]
    hotel_id_top5[i] = top5[i]
print(hotel_id_top5)
```
